### Question 1: Highest Oats producing country in 2010

```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
# opening the production dataset and viewing the first 5 rows

foodprod = pd.read_csv('Africa Food Production.csv')
foodprod.head()
```

```python
# Extracting oats data for the year 2010
oat_2010 = foodprod.query("Item == 'Oats' and Year == '2010'")
oat_2010
```

### Question 3: Top 3 highest rice producing countries

```python
# Extracting the dataset for rice production in African countries over the years
rice_df = foodprod.query("Item == 'Rice (Milled Equivalent)'")
 
# Finding the sum of values per country per year
rice_tot = pd.DataFrame(rice_df.groupby(['Country', 'Year'], as_index = False)['Value'].sum())
rice_tot
```

```python
# To obtain the top 3 countries with the highest rice production in kilotonnes over the years

top = pd.DataFrame(rice_tot.groupby(['Country', 'Year'], as_index = False)['Value'].sum())
top.sort_values(by = 'Value', ascending= False, inplace =True)
top.head(25)
```

> #### Answer: 
> Egypt, Nigeria, Madagascar.

### Question 4: Top 3 countries with the least food supply from 2004 to 2005 to 2006

```python
# Opening and viewing the first 5 rows of the food supply dataset

foodsup = pd.read_csv('Africa Food Supply.csv')
foodsup.head()
```

```python
# Extracting data for the year 2004
yr04 = foodsup.query("Year == '2004'")

# Grouping by total food Supply per Country
yr04_df = pd.DataFrame(yr04.groupby('Country', as_index = False)['Value'].sum())
yr04_df.sort_values(by = 'Value', ascending= True, inplace =True)
yr04_df
```

```python
# Extracting data for the year 2005
yr05 = foodsup.query("Year == '2005'")

# Grouping by total food Supply per Country
yr05_df = pd.DataFrame(yr05.groupby('Country', as_index = False)['Value'].sum())
yr05_df.sort_values(by = 'Value', ascending= True, inplace =True)
yr05_df
```

```python
# Extracting data for the year 2006
yr06 = foodsup.query("Year == '2006'")

# Grouping by total food Supply per Country
yr06_df = pd.DataFrame(yr06.groupby('Country', as_index = False)['Value'].sum())
yr06_df.sort_values(by = 'Value', ascending= True, inplace =True)
yr06_df
```

> #### Answer: 
> From the analysis of 3 years above, the most prominent countries with the least food supply are: Zambia, Ethiopia, Central African Republic.
